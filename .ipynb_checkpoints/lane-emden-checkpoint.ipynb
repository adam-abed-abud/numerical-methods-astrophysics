The aim of this project is to use numerical methods to solve the Lane-Emden equation.
This is a "dimensionless form of Poissonâ€™s equation for the gravitational potential of a Newtonian
self-gravitating, spherically symmetric, polytropic fluid"


$$\frac{1}{\xi^2}    \frac{d}{d\xi}  (\xi^2  \frac{d\theta}{d\xi}  )   = -\theta^n $$

where $\theta^n$ and $\xi$ are dimensionless variables
$\rho=\rho_c\theta^n$ and $r=\alpha\xi$
where $\alpha^2 = \frac{K(n+1)\rho^{\frac{1-n}{n}}_c}{4\pi G} $


Boundary conditions: $$\theta(\xi=0)=1$$, $$\frac{d\theta(\xi=0)}{d\xi}=0$$




Further info available in the pdf within this repo.

-----------------


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure



def analytical(t): #N=0
    return 1-t**2/6

def analytical2(t): #N=1
    return np.sin(t)/t

def analytical3(t): #N=5
    return (1+t**2/3)**(-0.5)

def f(r,u,theta,n):
    return -theta**n-2*u/r

index=[1.5,2,2.5,3,3.5] #polytropic index
```

Euler method -Forward

Simplest algorithm for solving ODE
Taylor expansion: $f(t+h)=f(t)+hy'(t)+\mathcal{O}(h^2)$ 

where $\mathcal{O}(h^2)$ is the order of the local truncation error for the step $y_n$  to $y_{n+1}$


Example: Starting at $y_0$ we calculate $y_1=y_0+hf(y_0)$
Global truncation error = $N\mathcal{O}(h^2)=\frac{L\mathcal{O}(h^2)}{h}=L\mathcal{O}(h)$ 


The smaller the step size the smaller the global error. not quite effective, double accuracy means double the number of points

```python

def euler_mthd( f, a, b, N, IV,V,polytropic_index):
    # determine step-size
    h = (b-a)/float(N)  

    # create mesh                         
    t = np.arange( a, b+h, h )
 
    # initialize w,z               
    w = np.zeros((N+1,)) 
    y = np.zeros((N+1,)) 

    # set initial values                     
    t[0], w[0] = IV  #theta
    t[0], y[0] = V #dtheta tcsi

    # apply Euler's method                       
    dr=a
    for i in range(1,N+1):       
        dr=dr+h
        w[i] = w[i-1] + h * y[i-1]
        y[i]=y[i-1]+h*f(dr,y[i-1],w[i],polytropic_index)
    plt.title( "Lane-Emden solutions with Euler's method")
    plt.legend(loc=4)
    axes = plt.gca()
    return w**polytropic_index
```

Multistep method

Also known as Adams-Bashford's method


Multistep method because it requires information from two previous steps

$y_{n+1}=y_n+h(\frac{3}{2}f(x_n,y_n)-\frac{1}{2}f(x_{n-1},y_{n-1}))+\mathcal{O}(h^3)$


```python

def multi( f, a, b, N, IV,V,polytropic_index):
 
    # determine step-size
    h = (b-a)/float(N)  

    # create mesh                         
    t = np.arange( a, b+h, h )
 
    # initialize w,z               
    w = np.zeros((N+1)) 
    y = np.zeros((N+2,)) 

    somma = np.zeros((N+1,)) #for benchmarking

    # set initial values                     
    t[0], w[0] = IV  #theta
    t[0], y[0] = V #dtheta tcsi

    # apply Euler's method                       
    dr=a
    for i in range(1,N+1): 
        dr=dr+h 
        w[i] = w[i-1] + h * y[i-1]
        y[i]=y[i-1]+h*f(dr,y[i-1],w[i],polytropic_index)
        f1=f( dr, y[i-1],w[i-1] ,polytropic_index)
        f2=f( dr,y[i-1], w[i-1]*h*f1,polytropic_index )
        y[i+1]=y[i]+h*(3*f1-f2)/2.0
        somma[i]=np.abs(analytical(t[i])-w[i])/analytical(t[i])	#N=0
    plt.title( "Lane-Emden solutions with multi step method")
    plt.xlabel(r'$\xi$', fontsize=20) 
    plt.ylabel(r'$\theta$', fontsize=20)
    plt.legend(loc=1)
    plt.grid()
    axes = plt.gca()
    return w
```

Runge-Kutta methods

Used to avoid calculating the derivative


Gives the best balance between speed and accuracy

$$k_1=hf(t_i,y_i) $$
$$k_2=hf(t_i+\frac{h}{2},y_i+\frac{k_1}{2}) $$
$$k_3=hf(t_i+\frac{h}{2},y_i+\frac{k_2}{2}) $$
$$k_4=hf(t_{i+1},y_i+k_3) $$

$$y_{i+1}=y_i+\frac{1}{6}(k_1+2k_2+2k_3+k_4))+\mathcal{O}(h^5) $$


```python
def runge( f, a, b, N, IV,V,polytropic_index):
 
    # determine step-size
    h = (b-a)/float(N)  

    # create mesh                         
    t = np.arange( a, b+h, h )
 
    # initialize w,z               
    w = np.zeros((N+1,)) 
    y = np.zeros((N+1,)) 

    somma = np.zeros((N+1,)) #for benchmarking

    # set initial values                     
    t[0], w[0] = IV  #theta
    t[0], y[0] = V #dtheta tcsi

    # apply Euler's method                       
    dr=a
    for i in range(1,N+1):       
        dr=dr+h
        w[i] = w[i-1] + h * y[i-1]
        k1 = f(dr, y[i-1],w[i-1],polytropic_index)
        k2 = f(dr+h/2.0,y[i-1], w[i-1]+0.5*k1*h,polytropic_index)
        k3 = f(dr+h/2.0,y[i-1] ,w[i-1]+0.5*k2*h,polytropic_index)
        k4 = f(dr+h, y[i-1], w[i-1]+k3*h,polytropic_index)
        y[i] = y[i-1]+(h/6.0)*(k1+2*k2+2*k3+k4)
        somma[i]=analytical(t[i])-w[i]#/analytical(t[i])	#N=0
    plt.plot( t, somma )
    plt.title( "Lane-Emden solutions with Runge-Kutta's method")
    plt.xlabel(r'$\xi$') 
    plt.ylabel(r'$\theta$')
    plt.legend(loc=4) 
    plt.grid()
    axes = plt.gca()
    return w
```

```python

a=0
b=10
N=100000

for i in range(len(index)):
    h = (b-a)/float(N)  
    t = np.arange( a, b+h, h )
    plt.plot(t,euler_mthd(f,0,10,100000,(0,1),(0,0),index[i]),label='n='+str(index[i]))
    
plt.xlabel(r'$\xi$',fontsize=20) 
plt.ylabel(r'$\theta$',fontsize=20)
axes = plt.gca()
axes.set_ylim([0,1]) #mi serve per N3
plt.legend(loc=1)
plt.grid()
plt.show()

    
    
for i in range(len(index)):
    h = (b-a)/float(N)  
    t = np.arange( a, b+h, h )
    plt.plot(t,multi(f,0,10,100000,(0,1),(0,0),index[i]),label='n='+str(index[i]))

plt.xlabel(r'$\xi$',fontsize=20) 
plt.ylabel(r'$\theta$',fontsize=20)
axes = plt.gca()
axes.set_ylim([0,1]) #mi serve per N3
plt.legend(loc=1)
plt.grid()
plt.show()


for i in range(len(index)):
    h = (b-a)/float(N)  
    t = np.arange( a, b+h, h )
    plt.plot(t,runge(f,0,10,100000,(0,1),(0,0),index[i]),label='n='+str(index[i]))

plt.xlabel(r'$\xi$',fontsize=20) 
plt.ylabel(r'$\theta$',fontsize=20)
axes = plt.gca()
axes.set_ylim([0,1]) #mi serve per N3
plt.legend(loc=1)
plt.grid()
plt.show()

```

Since there are analytical solutions for the Lane-Emden equation, it is possible to compare the performance of the various numerical methods. 

The analytical solutions correspond to three different polytropic indexes:


 n=0
 
 $         \theta(\xi)=1-\frac{1}{6}\xi^2$
 
 n=1
 
 $         \theta(\xi)=\frac{sin\xi}{\xi}$
 
 n=5
 
 $         \theta(\xi)=\frac{1}{\sqrt{1+\frac{\xi^2}{3}}}$


The following part is just an example on how to calculate the error of the Euler's method against the analytical solution for the polytropic index n=1. Further results can be found in the pdf within this repo.


```python
def euler_mthd_error( f, a, b, N, IV,V,polytropic_index):
 
    # determine step-size
    h = (b-a)/float(N)  

    # create mesh                         
    t = np.arange( a, b+h, h )
 
    # initialize w,z               
    w = np.zeros((N+1,)) 
    y = np.zeros((N+1,)) 

    difference = np.zeros((N+1,)) #for benchmarking

    # set initial values                     
    t[0], w[0] = IV  #theta
    t[0], y[0] = V #dtheta tcsi

    # apply Euler's method                       
    dr=a
    for i in range(1,N+1):       
        dr=dr+h
        w[i] = w[i-1] + h * y[i-1]
        y[i]=y[i-1]+h*f(dr,y[i-1],w[i],polytropic_index)
        difference[i]=analytical(t[i])-w[i]#/analytical(t[i])
    return difference

```

```python
a=0
b=10
N=100000

intervals=10

for i in range(4):
    intervals=10*intervals
    print("Step size {}: ".format(10/float(intervals))  )
    h = (b-a)/float(intervals) 
    x = np.linspace(0,10,intervals+1)  
    t = np.arange( a, b+h, h )
    plt.plot(x,euler_mthd_error(f,0,10,intervals,(0,1),(0,0), 1 ),label="Error w/ step-size "+str(10/float(intervals)))	

plt.title( "Lane-Emden scaling")
plt.xlabel('time') 
plt.ylabel('Error')
axes = plt.gca()
plt.legend(loc=2)
plt.show()

```
